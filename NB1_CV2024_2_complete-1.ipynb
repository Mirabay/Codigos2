```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage.color import rgb2gray, gray2rgb, rgb2lab, lab2rgb
from skimage.util import random_noise
from scipy.ndimage import uniform_filter, median_filter, gaussian_filter

#Not necessary only for jupyterthemes
# from jupyterthemes import jtplot
# jtplot.style()
```

```python
plots = lambda rows=1, cols=1, figsize=(20,10): plt.subplots(rows, cols, figsize=(figsize))
GREY = plt.cm.gray
REDS = plt.cm.Reds
GREENS = plt.cm.Greens
BLUES = plt.cm.Blues
```

#### Load a simple image

```python
lion = imread('images_class\lion.JPG')
lion_grey = rgb2gray(lion)
```

```python
lion.dtype
```

```python
f1, (ax1, ax2) = plots(1, 2)
ax1.imshow(lion), ax1.axis('off'), ax1.set_title('Colour lion', size=30)
ax2.imshow(lion_grey, cmap=GREY), ax2.axis('off'), ax2.set_title('Greyscale lion', size=30)
plt.show()
```
